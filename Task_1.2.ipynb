## Associate data1 with data2

Data1 contains the information of student number, and the records contained in data2 may contain multiple records of a single student number in data1, so the student number should be aggregated first.

After the aggregation, each student's "recent consumption to current interval (hours)" (Recent_consumption), "cumulative consumption times (take the maximum)" (Count_consumption), "total consumption amount" (Total_consumption) are obtained in the resulting DataFrame.

```python
#task 1.2
import numpy as np
import pandas as pd
```

```python
#task1_2_1.csv

# Read the table data after data1 cleaning
# Read the table data after data2 cleaning
formdata1=pd.read_csv("../newdata/task1_1_1.csv",encoding='gbk')
formdata2=pd.read_csv("../newdata/task1_1_2.csv",encoding='gbk')
```

```python
# Converts the card number to a character
formdata1["CardNo"]=formdata1["CardNo"].astype("str")
formdata2["CardNo"]=formdata1["CardNo"].astype("str")

# Converts the card number to a character
formdata2['Date'] = pd.to_datetime(formdata2['Date'])

# Screen out useful columns and consumption information about diet
newdata = formdata2.loc[:,['Type','Date','Money','FundMoney','CardNo','CardCount','Dept','Surplus']]

diet_data = newdata[(newdata['Dept'] == '第一食堂')|(newdata['Dept'] == '第二食堂')|\
                                (newdata['Dept'] == '第三食堂')|(newdata['Dept'] == '第四食堂')|\
                                (newdata['Dept'] == '第五食堂')|(newdata['Dept'] == '好利来食品店')]

diet_consumption_data = diet_data.groupby(by='Type').get_group('消费')

DataGroup =  diet_consumption_data.loc[:,['Date','Money','CardNo','CardCount']].groupby(by='CardNo')
```

```python
temp = pd.DataFrame()

temp['CardNo'] = [i[0] for i in DataGroup]
temp['temp'] = [np.max(i[1]['Date']) for i in DataGroup]
temp['Count_consumption'] = [np.max(i[1]['CardCount']) for i in DataGroup]
temp['Total_consumption'] = [np.sum(i[1]['Money']) for i in DataGroup]

temp['Recent_consumption'] = temp['temp'].apply(lambda x:(pd.Timestamp('2019-04-30 23:59:59')-x).seconds)
```

```python
#Associate the data1 and data2 tables
formdata1_formdata2 = pd.merge(formdata1,temp,left_on="CardNo",right_on="CardNo",how='inner')
formdata1_formdata2.to_csv("../result1/task1_2_1.csv",encoding='gbk')# Store the results of task 1.2
```
