```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data2 = pd.read_csv('../newdata/task1_1_2.csv',encoding='gbk')

data2['Date'] = pd.to_datetime(data2['Date'])
```

```python
##Extract the hour, add the hour column
data2['hour'] = [i.hour for i in data2['Date']]

#Filter out the canteen first, then overlap the time
data2_c1=data2[data2['Dept']=='第一食堂']
data2_c2=data2[data2['Dept']=='第二食堂']
data2_c3=data2[data2['Dept']=='第三食堂']
data2_c4=data2[data2['Dept']=='第四食堂']
data2_c5=data2[data2['Dept']=='第五食堂']
data2_c6=data2[data2['Dept']=='教师食堂']

##Calculate the number of people in each canteen
c1=len(data2_c1[['Date','CardNo','Money']].groupby(by = ['CardNo','Date']).count())
c2=len(data2_c2[['Date','CardNo','Money']].groupby(by = ['CardNo','Date']).count())
c3=len(data2_c3[['Date','CardNo','Money']].groupby(by = ['CardNo','Date']).count())
c4=len(data2_c4[['Date','CardNo','Money']].groupby(by = ['CardNo','Date']).count())
c5=len(data2_c5[['Date','CardNo','Money']].groupby(by = ['CardNo','Date']).count())
c6=len(data2_c6[['Date','CardNo','Money']].groupby(by = ['CardNo','Date']).count())


##Draw the pie chart
p1 = plt.figure(figsize=(7,7))
plt.rcParams['font.sans-serif'] = 'SimHei' ## Set Chinese display
plt.rcParams['axes.unicode_minus'] = False
colors = ['cornflowerblue','orange','limegreen','lightblue','aqua','pink']
plt.title("各个食堂用餐比例图",fontsize=16)
plt.pie([c1,c2,c3,c4,c5,c6],labels = ['第一食堂','第二食堂','第三食堂','第四食堂','第五食堂','教师食堂'],colors=colors,autopct='%1.1f%%')
plt.legend()
```

According to the pie chart, if the time factor is not considered. The number of dining restaurants from more to less in turn is: the second canteen, the fifth canteen, the fourth canteen, the first canteen and the third canteen.
Considering that the teachers' canteen is usually closed to students, it is not included in the scope of students' dining room.

```python
##Draw histogram of each restaurant at different time

mon_p = [mon_c1_s,mon_c2_s,mon_c3_s,mon_c4_s,mon_c5_s,mon_c6_s]
noon_p = [noon_c1_s,noon_c2_s,noon_c3_s,noon_c4_s,noon_c5_s,noon_c6_s]
night_p = [night_c1_s,night_c2_s,night_c3_s,night_c4_s,night_c5_s,night_c6_s]


p2 = plt.figure(figsize = (7,5))
plt.barh(range(6), mon_p,height=0.2, color='steelblue', alpha=0.8)      # Let me draw it from the bottom up
plt.yticks(range(6), ['第一食堂','第二食堂','第三食堂','第四食堂','第五食堂','教师食堂'],fontsize=14)
#plt.xlim(30,47)
plt.xlabel("人数",fontsize=14)
plt.title("不同餐厅各时段就餐人数",fontsize=14)
for x, y in enumerate(mon_p):
    plt.text(y + 0.2, x - 0.05, '%s' % y)
    
    
plt.barh([i+0.2 for i in range(6)], noon_p,height=0.2, color='indianred', alpha=0.8)  
for x, y in enumerate(noon_p):
    plt.text(y + 0.2, x +0.15, '%s' % y)

plt.barh([i-0.2 for i in range(6)], night_p,height=0.2, color='seagreen', alpha=0.8)  
for x, y in enumerate(night_p):
    plt.text(y + 0.2, x -0.25, '%s' % y)
plt.legend(['早','中','晚'])
```
